```python
from gensim.models import Word2Vec

sentences =[
    ["king", "queen", "man", "woman"],
    ["apple", "banana", "orange", "fruits"],
    ["king","man","ruler"],
    ["queen","woman","ruler"],
]

#train the model
model = Word2Vec(sentences,vector_size=3,window=2, min_count=1,sg=1)

#get the embedding for the king.
king = model.wv["king"]
print("getting the embedding for the king",king)

#find the similar words
similar_words = model.wv.most_similar("king")
print("similar words to king", similar_words)
```
