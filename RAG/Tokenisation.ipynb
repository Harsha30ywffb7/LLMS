Tokenisation mainly used in NLP to break the text into words and to anaylse, compare and we can find relationships and patterns behind it.

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize


text = "The quick brown fox jumps over the lazy dog."
words = word_tokenize(text)
sentences = sent_tokenize(text)

print(words)
print(sentences)
```

There are many tokenisers those are foundational models in NLP.
