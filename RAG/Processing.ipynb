```python
import pandas as pd

#initial data
reviews = [
    "OMG!!! I can't believe the Students were RUNNING soooo fast 😲🏃🏃‍♂️...",
"Yesterday, they visited 3 different Universities, Colleges & Institutes!!!",
"The PROFESSORS were teaching Machine-Learning, AI, and Data-Science; amazingly :)",
"Some students were coding, debugging, fixing ERRORS, and testing applications...",    
"Meanwhile, others were reading BOOKS 📚, writing NOTES!!!, and discussing projects???",
"The technologies included Python, Java, C++, JavaScript, React.js, and Node.js!!!",
"Finally, everyone said, 'We're learning better, faster, and smarter than ever before!!!' 🚀🔥"
]

#convert to dataframe
df_reviews = pd.DataFrame(reviews, columns=["Review"])
#convert to lowercase
df_reviews["Lowercase_reviews"] = df_reviews["Review"].str.lower()
#print
df_reviews


```

Removing punctuations and emojis

```python
import re
df_reviews["Review_noPunct"] = df_reviews["Lowercase_reviews"].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df_reviews
```

Remove stopwords for that take the nltk lib for corpus of the stopwords

```python


# import nltk
# from nltk.corpus import stopwords

# nltk.download('stopwords')

# stop_words = set(stopwords.words('english'))

stop_words = {
    "a", "an", "the", "is", "are", "was", "were",
    "in", "on", "at", "of", "to", "for", "and",
    "or", "but", "with", "by"
}
df_reviews["No_Stopwords"] = df_reviews["Review_noPunct"].apply(lambda x: ' '.join(word for word in x.split() if x not in stop_words))

df_reviews

```

Lemmatisation

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

df_reviews["Stemmed_reviews"] = df_reviews["No_Stopwords"].apply(lambda x: ' '.join(ps.stem(word) for word in x.split()))

df_reviews
```
