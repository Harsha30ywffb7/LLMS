```python
import os
from dotenv import load_dotenv
load_dotenv()
open_api_key = os.getenv("OPENAI_API_KEY")
```

```python
# interaction with the models
from langchain_openai import chatOpenAI
llm = chatOpenAI()
response = llm("Tell me the best ways to learn the AI and competetive programming")
print(response)
```

```python
#prompt templates, custom user inputs in messages
from langchain_core.prompts import PromptTemplate
prompt_template = PromptTemplate.from_template("Tell me the best ways to learn the {topic}")

prompt = prompt_template({"topic": "AI and competetive programming"})
response = llm(prompt)
print(response.content)
```

```python
movie_title_template = PromptTemplate.from_template(
    "suggest me more movies from the language {language} and genre {genre}"
)

prompt = movie_title_template({"language": "English", "genre": "Drama"})
response = llm(prompt)
print(response.content)
```

```python
#core and more important concept of chain of thought 
# just creating above tasks in a chain
movie_title_chain = movie_title_template | llm # here | is the pipe operator to chain

response = movie_title_chain.invoke({"language": "English", "genre": "Drama"})
print(response.content)

# this is for simple way of wrtiting code.
# now we can use the strOutputParser

from langchain.output_parsers import StrOutputParser
movie_title_chain = movie_title_template | llm | StrOutputParser()
response = movie_title_chain.invoke({"language": "English", "genre": "Drama"})

#now entire thing is done in single line
print(response.content)
```

```python
# creating multiple composed chains.
movie_summary_prompt = PromptTemplate.from_template("give me a summary of movie summary in 2-3 lines {movie_name}")

composed_chain = {"movie_name":movie_title_template} | movie_title_template | llm  | StrOutputParser()

composed_chain.invoke({"language": "English", "genre": "Drama"})
```

```python
#runnableLambdas:  a separate functions that are adding in the chain

from langchain_core.runnableLambdas import RunnableLambda

movie_name= RunnableLambda(lambda x: x["movie_name"])
movie_summary_prompt = PromptTemplate.from_template("give me a summary of movie summary in 2-3 lines {movie_name}")

composed_chain = movie_name | movie_summary_prompt | llm  | StrOutputParser()

composed_chain.invoke({"movie_name": "Avatar"})
```
